## 1. Import & Load Data

```python
# !pip install scikit-learn imbalanced-learn matplotlib seaborn pandas numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve, precision_recall_curve,
                             average_precision_score, ConfusionMatrixDisplay)
from sklearn.preprocessing import label_binarize
from sklearn.inspection import permutation_importance

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titlesize'] = 13
sns.set_style('whitegrid')
RS = 42
np.random.seed(RS)
print('Library siap!')
```

```python
df = pd.read_csv('gshs2015_for_classification.csv')
df = df.drop(columns=['mh_risk_score'])
TARGET    = 'target'
feat_cols = [c for c in df.columns if c != TARGET]
X_raw     = df[feat_cols].values
y         = df[TARGET].values

print(f'Dataset   : {df.shape[0]:,} baris × {df.shape[1]} kolom')
print(f'Fitur     : {len(feat_cols)}')
print(f'\nFitur:')
for i, c in enumerate(feat_cols, 1):
    print(f'  {i:2}. {c}')
print(f'\nDistribusi target:')
vc = pd.Series(y).value_counts().sort_index()
print(f'  0 (Tidak suicidal): {vc[0]:,} ({vc[0]/len(y)*100:.2f}%)')
print(f'  1 (Suicidal)       : {vc[1]:,} ({vc[1]/len(y)*100:.2f}%)')
print(f'  Rasio imbalance    : {vc[0]/vc[1]:.1f}:1')
```

---
## 2. Eksplorasi Awal & Class Imbalance

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
axes = axes.flatten()

key_feats = [
    'kesepian', 'khawatir_insomnia', 'hari_dibully', 'diserang_fisik',
    'flag_perokok_bin', 'flag_alkohol_bin', 'flag_narkoba_bin',
    'behavior_risk_score'
]

colors_cls = {0: '#A8D8A8', 1: '#FF6B6B'}

for i, col in enumerate(key_feats):
    ax = axes[i]
    is_categorical = df[col].nunique() <= 6
    
    sns.histplot(
        data=df, x=col, hue=TARGET, ax=ax, 
        palette=colors_cls, stat='density',
        common_norm=False, 
        discrete=is_categorical,
        multiple='dodge',
        alpha=0.7, bins=15 if not is_categorical else 'auto'
    )
    
    ax.set_title(col, fontsize=11, fontweight='bold')
    ax.set_xlabel('')
    ax.set_ylabel('Densitas', fontsize=9)

plt.suptitle('Distribusi Fitur Kunci per Kelas Target', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Korelasi fitur dengan target
corr_target = df.corr()[TARGET].drop(TARGET).sort_values()
fig, ax = plt.subplots(figsize=(8, 7))
colors_bar = ['#FF6B6B' if v > 0 else '#A8D8A8' for v in corr_target]
ax.barh(corr_target.index, corr_target.values, color=colors_bar, edgecolor='white')
ax.axvline(0, color='black', lw=0.8)
ax.set_title('Korelasi Setiap Fitur dengan Target (Suicidal Ideation)')
ax.set_xlabel('Korelasi Pearson')
plt.tight_layout()
plt.show()

print('Fitur paling berkorelasi positif dengan suicidal ideation:')
print(corr_target.tail(5).round(4).to_string())
```

---
## 3. Preprocessing untuk Modeling

### 3.1 Train-Test Split (Stratified)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.2, random_state=RS, stratify=y
)

print(f'Train : {len(X_train):,} ({len(X_train)/len(y)*100:.0f}%)')
print(f'  → Kelas 0: {(y_train==0).sum():,}  Kelas 1: {(y_train==1).sum():,}')
print(f'Test  : {len(X_test):,} ({len(X_test)/len(y)*100:.0f}%)')
print(f'  → Kelas 0: {(y_test==0).sum():,}  Kelas 1: {(y_test==1).sum():,}')
```

### 3.2 SMOTE — Oversampling Kelas Minoritas (pada data Training saja)

```python
try:
    from imblearn.over_sampling import SMOTE
    smote = SMOTE(random_state=RS)
    X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
    print(f'SMOTE berhasil:')
    print(f'  Sebelum → 0: {(y_train==0).sum():,}  1: {(y_train==1).sum():,}')
    print(f'  Sesudah → 0: {(y_train_sm==0).sum():,}  1: {(y_train_sm==1).sum():,}')
    USE_SMOTE = True
except ImportError:
    print('imbalanced-learn tidak terinstall → gunakan class_weight="balanced"')
    print('Install: pip install imbalanced-learn')
    X_train_sm, y_train_sm = X_train, y_train
    USE_SMOTE = False
```

```python
# Setup cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RS)
print('Stratified 5-Fold CV siap.')
```

```python
# Helper: evaluasi model
def evaluate_model(name, model, X_tr, y_tr, X_te, y_te, cv_obj):
    """Fit, predict, print classification report + confusion matrix + ROC."""
    model.fit(X_tr, y_tr)
    y_pred  = model.predict(X_te)
    y_proba = model.predict_proba(X_te)[:,1] if hasattr(model, 'predict_proba') else None

    print(f'\n{"="*55}')
    print(f'  {name}')
    print(f'{"="*55}')
    print(classification_report(y_te, y_pred, target_names=['Tidak(0)','Suicidal(1)'], digits=4))

    # CV F1
    cv_f1 = cross_val_score(model, X_raw, y, cv=cv_obj, scoring='f1', n_jobs=-1)
    print(f'CV F1 (5-fold): {cv_f1.mean():.4f} ± {cv_f1.std():.4f}')

    if y_proba is not None:
        auc = roc_auc_score(y_te, y_proba)
        ap  = average_precision_score(y_te, y_proba)
        print(f'ROC-AUC : {auc:.4f}')
        print(f'Avg Prec: {ap:.4f}')
    else:
        auc, ap = None, None

    # Confusion matrix
    fig, axes = plt.subplots(1, 2 if y_proba is not None else 1, figsize=(12, 4))
    if not isinstance(axes, np.ndarray): axes = [axes]
    ConfusionMatrixDisplay.from_predictions(
        y_te, y_pred, display_labels=['Tidak','Suicidal'],
        cmap='Blues', ax=axes[0])
    axes[0].set_title(f'Confusion Matrix — {name}')

    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_te, y_proba)
        axes[1].plot(fpr, tpr, lw=2, color='#E74C3C', label=f'AUC={auc:.4f}')
        axes[1].plot([0,1],[0,1], 'k--', lw=1)
        axes[1].set_title(f'ROC Curve — {name}')
        axes[1].set_xlabel('False Positive Rate')
        axes[1].set_ylabel('True Positive Rate')
        axes[1].legend()
    plt.tight_layout()
    plt.show()

    return model, {'auc': auc, 'ap': ap, 'y_pred': y_pred, 'y_proba': y_proba}

print('Helper function siap.')
```

---
## 4. Decision Tree (CART)

```python
# Tuning max_depth
depths = range(2, 11)
dt_f1_train, dt_f1_val = [], []
for d in depths:
    dt_tmp = DecisionTreeClassifier(max_depth=d, class_weight='balanced', random_state=RS)
    f1_cv  = cross_val_score(dt_tmp, X_train_sm, y_train_sm, cv=cv,
                              scoring='f1', n_jobs=-1)
    dt_f1_val.append(f1_cv.mean())
    dt_tmp.fit(X_train_sm, y_train_sm)
    from sklearn.metrics import f1_score
    dt_f1_train.append(f1_score(y_train_sm, dt_tmp.predict(X_train_sm)))

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(depths, dt_f1_train, 'o-', color='steelblue', label='Train F1')
ax.plot(depths, dt_f1_val,   's--', color='#E74C3C',  label='Val F1 (5-fold CV)')
ax.set_title('Decision Tree — Pengaruh max_depth terhadap F1')
ax.set_xlabel('max_depth'); ax.set_xticks(list(depths))
ax.set_ylabel('F1 Score (kelas 1)'); ax.legend()
plt.tight_layout()
plt.show()

best_depth = depths[np.argmax(dt_f1_val)]
print(f'Depth terbaik: {best_depth}  (val F1={max(dt_f1_val):.4f})')
```

```python
# Fit model final Decision Tree
dt_model = DecisionTreeClassifier(
    max_depth=best_depth,
    class_weight='balanced',
    criterion='gini',
    random_state=RS
)
dt_model, dt_metrics = evaluate_model(
    'Decision Tree', dt_model,
    X_train_sm, y_train_sm, X_test, y_test, cv
)
```

```python
# Visualisasi pohon (kedalaman 3 level saja agar terbaca)
fig, ax = plt.subplots(figsize=(20, 8))
plot_tree(dt_model, max_depth=3, feature_names=feat_cols,
          class_names=['Tidak','Suicidal'], filled=True,
          rounded=True, fontsize=8, ax=ax)
ax.set_title(f'Decision Tree (3 level pertama dari {best_depth} level)')
plt.tight_layout()
plt.show()

# Text rules (5 level)
print('Rules Decision Tree (5 level pertama):')
print(export_text(dt_model, feature_names=feat_cols, max_depth=5))
```

---
## 5. Random Forest

```python
# Tuning n_estimators
n_trees_list = [50, 100, 200, 300]
rf_auc_list  = []
for n in n_trees_list:
    rf_tmp = RandomForestClassifier(n_estimators=n, class_weight='balanced',
                                    random_state=RS, n_jobs=-1)
    auc_cv = cross_val_score(rf_tmp, X_train_sm, y_train_sm, cv=cv,
                              scoring='roc_auc', n_jobs=-1)
    rf_auc_list.append(auc_cv.mean())
    print(f'n_estimators={n:3d}: ROC-AUC={auc_cv.mean():.4f} ± {auc_cv.std():.4f}')

best_n = n_trees_list[np.argmax(rf_auc_list)]
print(f'\nJumlah pohon terbaik: {best_n}')
```

```python
# Fit model final Random Forest
rf_model = RandomForestClassifier(
    n_estimators=best_n,
    max_depth=None,
    class_weight='balanced',
    random_state=RS,
    n_jobs=-1
)
rf_model, rf_metrics = evaluate_model(
    'Random Forest', rf_model,
    X_train_sm, y_train_sm, X_test, y_test, cv
)
```

---
## 6. Support Vector Machine (SVM)

```python
# SVM dengan kernel RBF
# Catatan: SVM lambat pada dataset besar → gunakan subsample jika perlu
print('Fitting SVM (kernel=rbf, class_weight=balanced)...')

# Subsample training agar lebih cepat (opsional, hapus jika RAM cukup)
if len(X_train_sm) > 8000:
    np.random.seed(RS)
    idx_sub = np.random.choice(len(X_train_sm), 8000, replace=False)
    X_tr_svm, y_tr_svm = X_train_sm[idx_sub], y_train_sm[idx_sub]
    print(f'  Subsample training: 8000 dari {len(X_train_sm):,}')
else:
    X_tr_svm, y_tr_svm = X_train_sm, y_train_sm

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale',
                class_weight='balanced', probability=True, random_state=RS)
svm_model, svm_metrics = evaluate_model(
    'SVM (RBF)', svm_model,
    X_tr_svm, y_tr_svm, X_test, y_test, cv
)
```

---
## 7. Perbandingan Model

```python
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score

rows = []
for name, model, metrics in [
    ('Decision Tree',  dt_model,  dt_metrics),
    ('Random Forest',  rf_model,  rf_metrics),
    ('SVM (RBF)',       svm_model, svm_metrics),
]:
    yp = metrics['y_pred']
    rows.append({
        'Model'       : name,
        'Accuracy'    : round(accuracy_score(y_test, yp), 4),
        'Precision(1)': round(precision_score(y_test, yp, zero_division=0), 4),
        'Recall(1)'   : round(recall_score(y_test, yp, zero_division=0), 4),
        'F1(1)'       : round(f1_score(y_test, yp, zero_division=0), 4),
        'ROC-AUC'     : round(metrics['auc'], 4) if metrics['auc'] else '-',
        'Avg Precision': round(metrics['ap'], 4) if metrics['ap'] else '-',
    })

cmp_df = pd.DataFrame(rows).set_index('Model')
print('Perbandingan Model:')
cmp_df
```

```python
# Bar chart perbandingan
metrics_to_plot = ['Precision(1)', 'Recall(1)', 'F1(1)', 'ROC-AUC']
cmp_numeric = cmp_df[metrics_to_plot].apply(pd.to_numeric, errors='coerce')

x  = np.arange(len(metrics_to_plot))
w  = 0.25
models = cmp_numeric.index.tolist()
colors = ['#3498DB', '#2ECC71', '#E74C3C']

fig, ax = plt.subplots(figsize=(12, 5))
for i, (model, color) in enumerate(zip(models, colors)):
    offset = (i - len(models)/2 + 0.5) * w
    ax.bar(x + offset, cmp_numeric.loc[model], w,
           label=model, color=color, edgecolor='white', alpha=0.85)
    for xi, val in enumerate(cmp_numeric.loc[model]):
        ax.text(xi + offset, val + 0.005, f'{val:.3f}',
                ha='center', fontsize=8)

ax.set_xticks(x); ax.set_xticklabels(metrics_to_plot)
ax.set_ylabel('Score'); ax.set_ylim(0, 1.1)
ax.set_title('Perbandingan Kinerja Model Klasifikasi\n(Fokus pada kelas 1 = Suicidal)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Precision-Recall Curve semua model
fig, ax = plt.subplots(figsize=(8, 6))
for name, metrics, color in [
    ('Decision Tree', dt_metrics,  '#3498DB'),
    ('Random Forest', rf_metrics,  '#2ECC71'),
    ('SVM (RBF)',     svm_metrics, '#E74C3C'),
]:
    if metrics['y_proba'] is not None:
        prec, rec, _ = precision_recall_curve(y_test, metrics['y_proba'])
        ap = average_precision_score(y_test, metrics['y_proba'])
        ax.plot(rec, prec, lw=2, color=color, label=f'{name} (AP={ap:.3f})')

baseline = y_test.sum() / len(y_test)
ax.axhline(baseline, ls='--', color='grey', lw=1, label=f'Baseline={baseline:.3f}')
ax.set_title('Precision-Recall Curve')
ax.set_xlabel('Recall'); ax.set_ylabel('Precision')
ax.legend()
plt.tight_layout()
plt.show()
```

---
## 8. Feature Importance

```python
# Random Forest feature importance
fi_rf = pd.Series(rf_model.feature_importances_, index=feat_cols).sort_values(ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# RF Gini importance
colors_fi = ['#FF6B6B' if v > fi_rf.median() else '#5B9BD5' for v in fi_rf]
fi_rf.plot(kind='barh', ax=axes[0], color=colors_fi, edgecolor='white')
axes[0].axvline(fi_rf.median(), color='black', ls='--', lw=1, label='Median')
axes[0].set_title('Random Forest — Feature Importance (Gini)')
axes[0].set_xlabel('Importance')
axes[0].legend()

# Decision Tree importance
fi_dt = pd.Series(dt_model.feature_importances_, index=feat_cols).sort_values(ascending=True)
fi_dt.plot(kind='barh', ax=axes[1], color='#70AD47', edgecolor='white')
axes[1].set_title('Decision Tree — Feature Importance (Gini)')
axes[1].set_xlabel('Importance')

plt.tight_layout()
plt.show()

print('Top 5 fitur (Random Forest):')
print(fi_rf.sort_values(ascending=False).head(5).round(4))
```

```python
# Permutation importance (lebih reliable)
print('Menghitung permutation importance (Random Forest)...')
perm = permutation_importance(rf_model, X_test, y_test,
                              scoring='f1', n_repeats=10, random_state=RS)
perm_df = pd.Series(perm.importances_mean, index=feat_cols).sort_values()

fig, ax = plt.subplots(figsize=(9, 7))
colors_p = ['#FF6B6B' if v > 0 else '#A8A8A8' for v in perm_df]
perm_df.plot(kind='barh', ax=ax, color=colors_p, edgecolor='white')
ax.axvline(0, color='black', lw=0.8)
ax.set_title('Permutation Importance — Random Forest\n(Merah = berpengaruh positif terhadap F1)')
ax.set_xlabel('Mean Penurunan F1 saat Fitur Diacak')
plt.tight_layout()
plt.show()
```

## 9. (Opsional) Tambah Label Cluster sebagai Fitur

Bagian ini hanya dijalankan **setelah notebook clustering selesai** dan file `gshs2015_clustered_results.csv` tersedia.  
Jika clustering belum selesai, lewati bagian ini.

```python
import os
CLUSTER_FILE = 'gshs2015_clustered_results.csv'

if os.path.exists(CLUSTER_FILE):
    df_clust = pd.read_csv(CLUSTER_FILE)
    # Pastikan index selaras
    df_with_cluster = df.copy()
    df_with_cluster['cluster'] = df_clust['cluster'].values

    feat_cols_ext = feat_cols + ['cluster']
    X_ext = df_with_cluster[feat_cols_ext].values
    X_tr_ext, X_te_ext, y_tr_ext, y_te_ext = train_test_split(
        X_ext, y, test_size=0.2, random_state=RS, stratify=y)

    try:
        from imblearn.over_sampling import SMOTE
        X_tr_ext_sm, y_tr_ext_sm = SMOTE(random_state=RS).fit_resample(X_tr_ext, y_tr_ext)
    except:
        X_tr_ext_sm, y_tr_ext_sm = X_tr_ext, y_tr_ext

    rf_ext = RandomForestClassifier(n_estimators=best_n, class_weight='balanced',
                                    random_state=RS, n_jobs=-1)
    rf_ext.fit(X_tr_ext_sm, y_tr_ext_sm)
    yp_ext = rf_ext.predict(X_te_ext)

    print('Random Forest + Label Cluster:')
    print(classification_report(y_te_ext, yp_ext,
          target_names=['Tidak','Suicidal'], digits=4))
    print(f'ROC-AUC: {roc_auc_score(y_te_ext, rf_ext.predict_proba(X_te_ext)[:,1]):.4f}')
    print(f'\nBandingkan dengan RF tanpa cluster: F1={f1_score(y_test, rf_metrics["y_pred"]):.4f}')
else:
    print(f'File "{CLUSTER_FILE}" belum ada.')
    print('Jalankan notebook clustering.ipynb terlebih dahulu, lalu re-run cell ini.')
```
