```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

K_means

```python
df_cluster = pd.read_csv('gshs2015_for_clustering.csv')
if 'QN24' in df_cluster.columns:
    X = df_cluster.drop('QN24', axis=1)
else:
    X = df_cluster.copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"Bentuk data siap diolah: {X_scaled.shape}") 
```

```python
inertia = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

 
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(K_range, inertia, 'bo-', label='Inertia (Elbow)')
ax1.set_xlabel('Jumlah Klaster (K)')
ax1.set_ylabel('Inertia', color='b')
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()
ax2.plot(K_range, silhouette_scores, 'ro-', label='Silhouette Score')
ax2.set_ylabel('Silhouette Score', color='r')
ax2.tick_params(axis='y', labelcolor='r')

plt.title('Evaluasi Jumlah Klaster: Elbow vs Silhouette')
fig.tight_layout()
plt.show()
```

```python
OPTIMAL_K = 3 
kmeans_final = KMeans(n_clusters=OPTIMAL_K, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_scaled)

final_silhouette = silhouette_score(X_scaled, cluster_labels)
print(f"Final Silhouette Score (K={OPTIMAL_K}): {final_silhouette:.4f}")

df_cluster['Cluster_Label'] = cluster_labels

cluster_summary = df_cluster.groupby('Cluster_Label').mean()
print("\nKarakteristik tiap klaster (rata-rata fitur):")
display(cluster_summary)

print("\nSebaran jumlah siswa per klaster (K-Means, K=3):")
sebaran_kmeans = df_cluster['Cluster_Label'].value_counts().sort_index()

for cluster, jumlah in sebaran_kmeans.items():
    persentase = (jumlah / len(df_cluster)) * 100
    print(f"Cluster {cluster}: {jumlah} siswa ({persentase:.1f}%)")
```

HAC

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
df_cluster = pd.read_csv('gshs2015_for_clustering.csv')

if 'QN24' in df_cluster.columns:
    X = df_cluster.drop('QN24', axis=1)
else:
    X = df_cluster.copy()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
hag_model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
cluster_labels = hag_model.fit_predict(X_scaled)
```

```python

sil_score = silhouette_score(X_scaled, cluster_labels)
print(f"Final Silhouette Score (HAG - Ward, K=3): {sil_score:.4f}")

df_cluster['Cluster_Label'] = cluster_labels
print("\nSebaran jumlah siswa per klaster:")
sebaran = df_cluster['Cluster_Label'].value_counts().sort_index()

for cluster, jumlah in sebaran.items():
    persentase = (jumlah / len(df_cluster)) * 100
    print(f"Cluster {cluster}: {jumlah} siswa ({persentase:.1f}%)")
```

K-Medoid

```python
import os
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    f1_score, roc_auc_score, roc_curve, ConfusionMatrixDisplay,
    silhouette_score
)
from sklearn_extra.cluster import KMedoids
from xgboost import XGBClassifier
```

```python
RANDOM_STATE = 42
N_CLUSTERS   = 3
OUTPUT_DIR   = "output_visualisasi_model" 
DATA_PATH    = "gshs2015_for_clustering.csv"
os.makedirs(OUTPUT_DIR, exist_ok=True)

CLUSTER_NAMES  = [f"Cluster {i}" for i in range(N_CLUSTERS)]
CLUSTER_COLORS = ['#2196F3', '#FF9800', '#4CAF50']
MODEL_COLORS   = {
    "Decision Tree": '#E91E63',
    "Random Forest": '#9C27B0',
    "XGBoost":       '#FF5722',
}
```

```python
df = pd.read_csv(DATA_PATH)

stratify_col = df['QN24'] if 'QN24' in df.columns else None
drop_cols = ['QN24'] if 'QN24' in df.columns else []
X_all = df.drop(columns=drop_cols)
X_all = df.drop(columns=['mh_risk_score', 'bmi_encoded', 'jenjang_bin', 'jumlah_teman', 'perkelahian_fisik',
                      'flag_aktif_fisik_bin', 'behavior_risk_score', 'kelas'])
feature_names = X_all.columns.tolist()

print(f"Total sampel  : {len(X_all)}")
print(f"Jumlah fitur  : {len(feature_names)}")
print(f"Stratify by   : {'QN24' if stratify_col is not None else 'None'}")
```

```python
idx = np.arange(len(X_all))
idx_train, idx_test = train_test_split(
    idx, test_size=0.2, random_state=RANDOM_STATE, stratify=stratify_col
)

X_train_raw = X_all.iloc[idx_train].values
X_test_raw  = X_all.iloc[idx_test].values
print(f"Train: {len(idx_train)} sampel | Test: {len(idx_test)} sampel")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_raw)   
X_test_scaled  = scaler.transform(X_test_raw)         

kmedoids = KMedoids(
    n_clusters=N_CLUSTERS,
    metric='manhattan',
    init='k-medoids++',
    random_state=RANDOM_STATE
)
kmedoids.fit(X_train_scaled)

y_train = kmedoids.labels_
y_test  = kmedoids.predict(X_test_scaled)   

print("Distribusi klaster (train):")
for cl in range(N_CLUSTERS):
    n = (y_train == cl).sum()
    print(f"  Cluster {cl}: {n:5d} ({n/len(y_train)*100:.1f}%)")

print("Distribusi klaster (test):")
for cl in range(N_CLUSTERS):
    n = (y_test == cl).sum()
    print(f"  Cluster {cl}: {n:5d} ({n/len(y_test)*100:.1f}%)")

sil_score = silhouette_score(X_train_scaled, y_train, metric='manhattan')
print(f"Final Silhouette Score (K-Medoids Train Set, K={N_CLUSTERS}): {sil_score:.4f}")
```

```python
# Visualisasi PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y_train, 
                palette='viridis', alpha=0.5, s=20)
plt.title('Sebaran Klaster K-Medoids Train Set (Proyeksi PCA 2D)')
plt.xlabel(f'Komponen Utama 1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'Komponen Utama 2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
plt.legend(title='Klaster')
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/fig_clustering_pca.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
df_train_eval = pd.DataFrame(X_train_raw, columns=feature_names)
df_train_eval['Cluster_Label'] = y_train
cluster_means = df_train_eval.groupby('Cluster_Label').mean()

plt.figure(figsize=(10, 8))
sns.heatmap(cluster_means.T, annot=True, fmt=".2f", cmap="YlOrRd", linewidths=.5)
plt.title('Heatmap Karakteristik Rata-rata per Klaster (Train Set)')
plt.xlabel('Cluster')
plt.ylabel('Fitur')
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/fig_clustering_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
```

Klasifikasi

```python
models = {
    "Decision Tree": DecisionTreeClassifier(
        max_depth=8, min_samples_leaf=20, random_state=RANDOM_STATE
    ),
    "Random Forest": RandomForestClassifier(
        n_estimators=200, max_depth=10, min_samples_leaf=10,
        n_jobs=-1, random_state=RANDOM_STATE
    ),
    "XGBoost": XGBClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.1,
        subsample=0.8, colsample_bytree=0.8,
        eval_metric='mlogloss', random_state=RANDOM_STATE,
        verbosity=0
    ),
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
results = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_prob = model.predict_proba(X_test_scaled)
    
    acc    = accuracy_score(y_test, y_pred)
    f1_mac = f1_score(y_test, y_pred, average='macro')
    f1_wei = f1_score(y_test, y_pred, average='weighted')
    auc    = roc_auc_score(y_test, y_prob, multi_class='ovr', average='macro')
    
    cv_acc = cross_val_score(
        model, X_train_scaled, y_train, cv=cv, scoring='accuracy', n_jobs=-1
    ).mean()
    
    results[name] = dict(
        model=model, y_pred=y_pred, y_prob=y_prob,
        accuracy=acc, f1_macro=f1_mac, f1_weighted=f1_wei,
        auc_macro=auc, cv_acc=cv_acc
    )
    
    print(f"\n[{name}]")
    print(f"  Accuracy        : {acc:.4f}")
    print(f"  F1 Macro        : {f1_mac:.4f}")
    print(f"  AUC (OvR macro) : {auc:.4f}")
    print(f"  CV Accuracy     : {cv_acc:.4f}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Confusion Matrix — RF | DT | XGBoost', fontsize=13, fontweight='bold', y=1.02)
for ax, (name, res) in zip(axes, results.items()):
    cm = confusion_matrix(y_test, res["y_pred"])
    ConfusionMatrixDisplay(cm, display_labels=CLUSTER_NAMES).plot(ax=ax, colorbar=False, cmap='Blues', values_format='d')
    ax.set_title(f"{name}\nAcc={res['accuracy']:.4f} | F1={res['f1_macro']:.4f}", fontsize=10, fontweight='bold', color=MODEL_COLORS[name])
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/fig3_confusion_matrix.png", dpi=150)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Feature Importance — Top 15 Fitur per Model', fontsize=13, fontweight='bold', y=1.01)
for ax, (name, res) in zip(axes, results.items()):
    imp = res["model"].feature_importances_
    idx_top = np.argsort(imp)[-15:]
    vals = imp[idx_top]
    feats = [feature_names[i] for i in idx_top]
    bars = ax.barh(range(len(idx_top)), vals, color=MODEL_COLORS[name], alpha=0.8)
    ax.set_yticks(range(len(idx_top)))
    ax.set_yticklabels(feats, fontsize=8)
    ax.set_title(name, fontsize=11, fontweight='bold', color=MODEL_COLORS[name])
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/fig4_feature_importance.png", dpi=150)
plt.show()
```

```python
df_bridge = pd.DataFrame({
    'Cluster': y_train, 
    'Ide_Bunuh_Diri': stratify_col.iloc[idx_train].values if stratify_col is not None else np.zeros(len(y_train))
})
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df_bridge, x='Cluster', hue='Ide_Bunuh_Diri', palette='Set1')
plt.title('Distribusi Aktual Ide Bunuh Diri (QN24) per Klaster K-Medoids')
plt.xlabel('Klaster')
plt.ylabel('Jumlah Siswa')
plt.legend(title='Ide Bunuh Diri (0=Tidak, 1=Ya)')
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                    ha='center', va='bottom', fontsize=9, xytext=(0, 3), textcoords='offset points')
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/fig5_cluster_vs_target.png", dpi=150)
plt.show()
```

```python
metrics_df = pd.DataFrame({
    name: [res['accuracy'], res['f1_macro'], res['auc_macro']]
    for name, res in results.items()
}, index=['Accuracy', 'F1-Macro', 'AUC']).T

metrics_df.plot(kind='bar', figsize=(9, 6), colormap='viridis', alpha=0.85)
plt.title('Perbandingan Performa Algoritma Klasifikasi', fontweight='bold')
plt.ylabel('Skor Metrik')
plt.xlabel('Algoritma Model')
plt.xticks(rotation=0)
plt.ylim(0, 1.15)
plt.legend(loc='lower right', title='Metrik Evaluasi')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/fig6_model_comparison.png", dpi=150)
plt.show()
```
