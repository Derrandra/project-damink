## 2. Feature Engineering

Menambahkan fitur interaksi dan komposit untuk memperkuat sinyal prediksi.  
Fitur baru dibuat berdasarkan domain knowledge kesehatan mental remaja.

> **Penting:** Tidak menggunakan `suicidal_ideation_raw`, `rencana_bunuh_diri`,
> `percobaan_bunuh_diri`, atau `mh_risk_score` karena merupakan **data leakage**
> (variabel tersebut langsung turunan dari target).

```python
df = df_raw.copy()

# ── Fitur interaksi (perkalian dua variabel berisiko) ─────────────────
df['interaksi_kesepian_bully']    = df['kesepian'] * df['hari_dibully']
df['interaksi_khawatir_bully']    = df['khawatir_insomnia'] * df['hari_dibully']
df['interaksi_khawatir_diserang'] = df['khawatir_insomnia'] * df['diserang_fisik']
df['interaksi_kesepian_diserang'] = df['kesepian'] * df['diserang_fisik']

# ── Fitur komposit ────────────────────────────────────────────────────
# Skor tekanan psikologis (kesepian + khawatir + bully)
df['skor_tekanan']   = df['kesepian'] + df['khawatir_insomnia'] + df['hari_dibully']

# Total penggunaan zat
df['jumlah_zat']     = df['flag_perokok_bin'] + df['flag_alkohol_bin'] + df['flag_narkoba_bin']

# Isolasi sosial: jumlah teman <= 2 (kode 1 atau 2 dari Q27)
df['isolasi_sosial'] = (df['jumlah_teman'] <= 2).astype(int)

# Risiko ganda: kesepian TINGGI (>=3) DAN di-bully (>=2 hari)
df['risiko_ganda']   = ((df['kesepian'] >= 3) & (df['hari_dibully'] >= 2)).astype(int)

# Kekerasan gabungan (diserang + perkelahian)
df['total_kekerasan'] = df['diserang_fisik'] + df['perkelahian_fisik']

feat_cols = [c for c in df.columns if c != TARGET]
print(f'Total fitur setelah feature engineering: {len(feat_cols)}')
print(f'\\nFitur baru yang ditambahkan:')
new_feats = ['interaksi_kesepian_bully', 'interaksi_khawatir_bully',
             'interaksi_khawatir_diserang', 'interaksi_kesepian_diserang',
             'skor_tekanan', 'jumlah_zat', 'isolasi_sosial',
             'risiko_ganda', 'total_kekerasan']
for f in new_feats:
    print(f'  + {f}')
```

```python
# Korelasi fitur dengan target
corr_target = df.corr()[TARGET].drop(TARGET).sort_values()
fig, ax = plt.subplots(figsize=(8, 7))
colors_bar = ['#FF6B6B' if v > 0 else '#A8D8A8' for v in corr_target]
ax.barh(corr_target.index, corr_target.values, color=colors_bar, edgecolor='white')
ax.axvline(0, color='black', lw=0.8)
ax.set_title('Korelasi Fitur dengan Target (Suicidal Ideation)')
ax.set_xlabel('Korelasi Pearson')
for i, (idx, val) in enumerate(corr_target.items()):
    ax.text(val + (0.002 if val >= 0 else -0.002), i,
            f'{val:.3f}', va='center', fontsize=8,
            ha='left' if val >= 0 else 'right')
plt.tight_layout()
plt.show()

print('Top 5 fitur berkorelasi positif:')
print(corr_target.tail(5).round(4).to_string())
```

## 4. Perbandingan Strategi Resampling

Membandingkan 4 teknik resampling untuk menangani imbalanced data.  
Evaluasi menggunakan **Random Forest** sebagai baseline model.

| Teknik | Cara Kerja |
|---|---|
| **SMOTE** | Interpolasi linier antar sampel minoritas |
| **BorderlineSMOTE** | Fokus pada sampel minoritas yang berada di *borderline* (paling sulit diklasifikasikan) |
| **ADASYN** | Oversample adaptif — lebih banyak sintetis di area yang sulit |
| **SMOTETomek** | SMOTE + hapus sampel mayoritas yang terlalu dekat (under+over) |

```python
samplers = {
    'SMOTE'          : SMOTE(random_state=RS),
    'BorderlineSMOTE': BorderlineSMOTE(random_state=RS),
    'ADASYN'         : ADASYN(random_state=RS),
    'SMOTETomek'     : SMOTETomek(random_state=RS),
}

resampling_results = {}
print(f'{"Sampler":20s}  {"n_train":>8}  {"F1(1)":>7}  {"AUC":>7}  {"Prec":>7}  {"Recall":>7}')
print('-' * 62)

for name, sampler in samplers.items():
    try:
        X_res, y_res = sampler.fit_resample(X_train, y_train)
        rf_tmp = RandomForestClassifier(n_estimators=200, class_weight='balanced',
                                        random_state=RS, n_jobs=-1)
        rf_tmp.fit(X_res, y_res)
        yp  = rf_tmp.predict(X_test)
        ypr = rf_tmp.predict_proba(X_test)[:, 1]
        f1  = f1_score(y_test, yp)
        auc = roc_auc_score(y_test, ypr)
        pre = precision_score(y_test, yp, zero_division=0)
        rec = recall_score(y_test, yp)
        resampling_results[name] = {
            'X_res': X_res, 'y_res': y_res,
            'f1': f1, 'auc': auc, 'precision': pre, 'recall': rec
        }
        print(f'{name:20s}  {len(X_res):>8,}  {f1:>7.4f}  {auc:>7.4f}  {pre:>7.4f}  {rec:>7.4f}')
    except Exception as e:
        print(f'{name:20s}: ERROR — {e}')

# Pilih sampler terbaik (F1)
BEST_SAMPLER = max(resampling_results, key=lambda k: resampling_results[k]['f1'])
X_train_res  = resampling_results[BEST_SAMPLER]['X_res']
y_train_res  = resampling_results[BEST_SAMPLER]['y_res']
print(f'\n→ Sampler terpilih: {BEST_SAMPLER}  (F1={resampling_results[BEST_SAMPLER]["f1"]:.4f})')
```

```python
# Visualisasi perbandingan sampler
metrics_smp = ['f1', 'auc', 'precision', 'recall']
labels_smp  = ['F1 (kelas 1)', 'ROC-AUC', 'Precision', 'Recall']
x  = np.arange(len(metrics_smp))
w  = 0.8 / len(resampling_results)
colors_smp = sns.color_palette('tab10', len(resampling_results))

fig, ax = plt.subplots(figsize=(13, 5))
for i, (name, res) in enumerate(resampling_results.items()):
    offset = (i - len(resampling_results)/2 + 0.5) * w
    vals = [res[m] for m in metrics_smp]
    bars = ax.bar(x + offset, vals, w, label=name,
                  color=colors_smp[i], edgecolor='white', alpha=0.85)
    for bar, val in zip(bars, vals):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
                f'{val:.3f}', ha='center', fontsize=7, rotation=90)

ax.set_xticks(x); ax.set_xticklabels(labels_smp)
ax.set_ylabel('Score'); ax.set_ylim(0, 1.0)
ax.set_title('Perbandingan Strategi Resampling\n(Model: Random Forest 200 trees)')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()

# Distribusi kelas setelah resampling
print('Distribusi kelas setelah resampling:')
for name, res in resampling_results.items():
    c0 = (res['y_res']==0).sum()
    c1 = (res['y_res']==1).sum()
    print(f'  {name:20s}: 0={c0:,}  1={c1:,}  total={len(res["y_res"]):,}')
```

```python
# Helper evaluasi model
def evaluate_model(name, model, X_tr, y_tr, X_te, y_te, threshold=0.5):
    model.fit(X_tr, y_tr)
    y_proba = model.predict_proba(X_te)[:, 1]
    y_pred  = (y_proba >= threshold).astype(int)

    print(f'\n{"="*58}')
    print(f'  {name}  (threshold={threshold})')
    print(f'{"="*58}')
    print(classification_report(y_te, y_pred,
          target_names=['Tidak(0)', 'Suicidal(1)'], digits=4))

    auc = roc_auc_score(y_te, y_proba)
    ap  = average_precision_score(y_te, y_proba)
    print(f'ROC-AUC         : {auc:.4f}')
    print(f'Avg Precision   : {ap:.4f}')

    fig, axes = plt.subplots(1, 2, figsize=(13, 4))

    ConfusionMatrixDisplay.from_predictions(
        y_te, y_pred, display_labels=['Tidak', 'Suicidal'],
        cmap='Blues', ax=axes[0])
    axes[0].set_title(f'Confusion Matrix — {name}')

    fpr, tpr, _ = roc_curve(y_te, y_proba)
    axes[1].plot(fpr, tpr, lw=2, color='#E74C3C', label=f'AUC={auc:.4f}')
    axes[1].plot([0,1], [0,1], 'k--', lw=1)
    axes[1].set_title(f'ROC Curve — {name}')
    axes[1].set_xlabel('False Positive Rate')
    axes[1].set_ylabel('True Positive Rate')
    axes[1].legend()
    plt.tight_layout()
    plt.show()

    return model, {'auc': auc, 'ap': ap, 'y_pred': y_pred, 'y_proba': y_proba, 'threshold': threshold}

print('Helper function siap.')
```

## 5. Decision Tree (CART)

```python
# Tuning max_depth via CV
depths = range(2, 12)
dt_val_f1 = []
for d in depths:
    dt_tmp = DecisionTreeClassifier(max_depth=d, class_weight='balanced', random_state=RS)
    f1_cv  = cross_val_score(dt_tmp, X_train_res, y_train_res,
                              cv=cv, scoring='f1', n_jobs=-1)
    dt_val_f1.append(f1_cv.mean())
    print(f'  depth={d:2d}: CV F1={f1_cv.mean():.4f} ± {f1_cv.std():.4f}')

best_depth = list(depths)[np.argmax(dt_val_f1)]

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(depths, dt_val_f1, 'o-', color='steelblue', lw=2, ms=7)
ax.axvline(best_depth, color='red', ls='--', alpha=0.7, label=f'Best depth={best_depth}')
ax.set_title('Decision Tree — CV F1 vs max_depth')
ax.set_xlabel('max_depth'); ax.set_xticks(list(depths))
ax.set_ylabel('CV F1 Score (kelas 1)'); ax.legend()
plt.tight_layout(); plt.show()
print(f'\nDepth terbaik: {best_depth}  (CV F1={max(dt_val_f1):.4f})')
```

```python
dt_model = DecisionTreeClassifier(
    max_depth=best_depth, class_weight='balanced',
    criterion='gini', random_state=RS
)
dt_model, dt_metrics = evaluate_model(
    'Decision Tree', dt_model,
    X_train_res, y_train_res, X_test, y_test
)
```

```python
# Visualisasi pohon (3 level)
fig, ax = plt.subplots(figsize=(20, 8))
plot_tree(dt_model, max_depth=3, feature_names=feat_cols,
          class_names=['Tidak', 'Suicidal'],
          filled=True, rounded=True, fontsize=8, ax=ax)
ax.set_title(f'Decision Tree — 3 Level Pertama (dari total {best_depth} level)')
plt.tight_layout(); plt.show()

print('Rules 4 level pertama:')
print(export_text(dt_model, feature_names=feat_cols, max_depth=4))
```

## 6. Random Forest

```python
# Tuning n_estimators
n_trees_list = [50, 100, 200, 300, 500]
rf_cv_auc = []
for n in n_trees_list:
    rf_tmp = RandomForestClassifier(n_estimators=n, class_weight='balanced',
                                    random_state=RS, n_jobs=-1)
    auc_cv = cross_val_score(rf_tmp, X_train_res, y_train_res,
                              cv=cv, scoring='roc_auc', n_jobs=-1)
    rf_cv_auc.append(auc_cv.mean())
    print(f'  n={n:3d}: CV AUC={auc_cv.mean():.4f} ± {auc_cv.std():.4f}')

best_n = n_trees_list[np.argmax(rf_cv_auc)]

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(n_trees_list, rf_cv_auc, 'o-', color='#2ECC71', lw=2, ms=7)
ax.axvline(best_n, color='red', ls='--', alpha=0.7, label=f'Best n={best_n}')
ax.set_title('Random Forest — CV ROC-AUC vs n_estimators')
ax.set_xlabel('n_estimators'); ax.set_ylabel('CV ROC-AUC')
ax.legend(); plt.tight_layout(); plt.show()
print(f'\nJumlah pohon terbaik: {best_n}  (CV AUC={max(rf_cv_auc):.4f})')
```

```python
rf_model = RandomForestClassifier(
    n_estimators=best_n, class_weight='balanced',
    random_state=RS, n_jobs=-1
)
rf_model, rf_metrics = evaluate_model(
    'Random Forest', rf_model,
    X_train_res, y_train_res, X_test, y_test
)
```

## 7. XGBoost

Menggunakan `scale_pos_weight` untuk menangani imbalance secara internal.

```python
# Tuning n_estimators & learning_rate
xgb_configs = [
    {'n_estimators': 100,  'learning_rate': 0.1,  'max_depth': 5},
    {'n_estimators': 200,  'learning_rate': 0.1,  'max_depth': 5},
    {'n_estimators': 300,  'learning_rate': 0.05, 'max_depth': 5},
    {'n_estimators': 300,  'learning_rate': 0.1,  'max_depth': 6},
]

print(f'{"n_est":>6}  {"lr":>6}  {"depth":>6}  {"CV AUC":>9}')
print('-' * 36)
xgb_cv_results = []
for cfg in xgb_configs:
    xgb_tmp = XGBClassifier(
        **cfg, scale_pos_weight=SCALE_POS,
        random_state=RS, eval_metric='logloss', verbosity=0, n_jobs=-1
    )
    auc_cv = cross_val_score(xgb_tmp, X_train_res, y_train_res,
                              cv=cv, scoring='roc_auc', n_jobs=-1)
    xgb_cv_results.append(auc_cv.mean())
    print(f'{cfg["n_estimators"]:>6}  {cfg["learning_rate"]:>6}  {cfg["max_depth"]:>6}  {auc_cv.mean():>9.4f} ± {auc_cv.std():.4f}')

best_xgb_cfg = xgb_configs[np.argmax(xgb_cv_results)]
print(f'\nKonfigurasi terbaik: {best_xgb_cfg}')
```

```python
xgb_model = XGBClassifier(
    **best_xgb_cfg,
    scale_pos_weight=SCALE_POS,
    random_state=RS,
    eval_metric='logloss',
    verbosity=0,
    n_jobs=-1
)
xgb_model, xgb_metrics = evaluate_model(
    'XGBoost', xgb_model,
    X_train_res, y_train_res, X_test, y_test
)
```

## 8. Threshold Tuning

Default threshold 0.5 biasanya terlalu tinggi untuk kelas minoritas yang sangat kecil.  
Kita cari threshold optimal yang memaksimalkan **F1-score kelas 1**.

> **Catatan:** Threshold tuning hanya boleh dilakukan pada **validation set / CV**, bukan test set.  
> Di sini kita tampilkan kurva pada test set untuk keperluan visualisasi.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
thresholds = np.linspace(0.05, 0.7, 100)

best_thresholds = {}
for ax, (name, metrics, color) in zip(axes, [
    ('Decision Tree', dt_metrics,  '#3498DB'),
    ('Random Forest', rf_metrics,  '#2ECC71'),
    ('XGBoost',       xgb_metrics, '#E74C3C'),
]):
    proba = metrics['y_proba']
    f1s   = [f1_score(y_test, (proba >= t).astype(int), zero_division=0) for t in thresholds]
    pres  = [precision_score(y_test, (proba >= t).astype(int), zero_division=0) for t in thresholds]
    recs  = [recall_score(y_test, (proba >= t).astype(int), zero_division=0) for t in thresholds]

    best_t  = thresholds[np.argmax(f1s)]
    best_f1 = max(f1s)
    best_thresholds[name] = best_t

    ax.plot(thresholds, f1s,  lw=2, color=color,     label=f'F1')
    ax.plot(thresholds, pres, lw=1.5, color='#F39C12', ls='--', label='Precision')
    ax.plot(thresholds, recs, lw=1.5, color='#8E44AD', ls='--', label='Recall')
    ax.axvline(best_t, color='black', ls=':', lw=1.5,
               label=f'Best t={best_t:.2f}\nF1={best_f1:.4f}')
    ax.set_title(f'Threshold Tuning — {name}')
    ax.set_xlabel('Threshold'); ax.set_ylabel('Score')
    ax.legend(fontsize=8); ax.set_xlim(0.05, 0.7)

plt.suptitle('Pencarian Threshold Optimal (Maks. F1 kelas 1)', fontsize=13)
plt.tight_layout(); plt.show()

print('Threshold optimal per model:')
for name, t in best_thresholds.items():
    proba = {'Decision Tree': dt_metrics, 'Random Forest': rf_metrics,
             'XGBoost': xgb_metrics}[name]['y_proba']
    yp    = (proba >= t).astype(int)
    print(f'  {name:15s}: threshold={t:.2f}  F1={f1_score(y_test,yp):.4f}  '
          f'Prec={precision_score(y_test,yp,zero_division=0):.4f}  '
          f'Recall={recall_score(y_test,yp):.4f}')
```

```python
# Re-evaluasi setiap model dengan threshold optimal
print('Evaluasi ulang dengan threshold optimal:')
dt_model_t, dt_metrics_t   = evaluate_model(
    f'Decision Tree (t={best_thresholds["Decision Tree"]:.2f})',
    dt_model, X_train_res, y_train_res, X_test, y_test,
    threshold=best_thresholds['Decision Tree']
)
rf_model_t, rf_metrics_t   = evaluate_model(
    f'Random Forest (t={best_thresholds["Random Forest"]:.2f})',
    rf_model, X_train_res, y_train_res, X_test, y_test,
    threshold=best_thresholds['Random Forest']
)
xgb_model_t, xgb_metrics_t = evaluate_model(
    f'XGBoost (t={best_thresholds["XGBoost"]:.2f})',
    xgb_model, X_train_res, y_train_res, X_test, y_test,
    threshold=best_thresholds['XGBoost']
)
```

```python
rows = []
for name, metrics in [
    ('Decision Tree',  dt_metrics_t),
    ('Random Forest',  rf_metrics_t),
    ('XGBoost',        xgb_metrics_t),
]:
    yp = metrics['y_pred']
    rows.append({
        'Model'        : name,
        'Threshold'    : round(metrics['threshold'], 2),
        'Accuracy'     : round(accuracy_score(y_test, yp), 4),
        'Precision(1)' : round(precision_score(y_test, yp, zero_division=0), 4),
        'Recall(1)'    : round(recall_score(y_test, yp, zero_division=0), 4),
        'F1(1)'        : round(f1_score(y_test, yp, zero_division=0), 4),
        'ROC-AUC'      : round(metrics['auc'], 4),
        'Avg Precision': round(metrics['ap'], 4),
    })

cmp_df = pd.DataFrame(rows).set_index('Model')
print('Perbandingan Model (dengan threshold optimal):')
cmp_df
```

```python
# Bar chart perbandingan metrik utama
metrics_plot = ['Precision(1)', 'Recall(1)', 'F1(1)', 'ROC-AUC']
cmp_num = cmp_df[metrics_plot].apply(pd.to_numeric, errors='coerce')
x = np.arange(len(metrics_plot)); w = 0.25
colors_m = ['#3498DB', '#2ECC71', '#E74C3C']

fig, ax = plt.subplots(figsize=(13, 5))
for i, (model, color) in enumerate(zip(cmp_num.index, colors_m)):
    offset = (i - 1) * w
    bars = ax.bar(x + offset, cmp_num.loc[model], w,
                  label=model, color=color, edgecolor='white', alpha=0.85)
    for bar, val in zip(bars, cmp_num.loc[model]):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
                f'{val:.3f}', ha='center', fontsize=8)

ax.set_xticks(x); ax.set_xticklabels(metrics_plot)
ax.set_ylabel('Score'); ax.set_ylim(0, 1.1)
ax.set_title('Perbandingan Model Klasifikasi\n(threshold optimal masing-masing model)')
ax.legend(); plt.tight_layout(); plt.show()
```

```python
# Precision-Recall & ROC Curve overlay
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
model_specs = [
    ('Decision Tree', dt_metrics_t,  '#3498DB'),
    ('Random Forest', rf_metrics_t,  '#2ECC71'),
    ('XGBoost',       xgb_metrics_t, '#E74C3C'),
]
for name, metrics, color in model_specs:
    proba = metrics['y_proba']
    # ROC
    fpr, tpr, _ = roc_curve(y_test, proba)
    auc = roc_auc_score(y_test, proba)
    axes[0].plot(fpr, tpr, lw=2, color=color, label=f'{name} (AUC={auc:.4f})')
    # PR
    prec, rec, _ = precision_recall_curve(y_test, proba)
    ap = average_precision_score(y_test, proba)
    axes[1].plot(rec, prec, lw=2, color=color, label=f'{name} (AP={ap:.4f})')

axes[0].plot([0,1],[0,1], 'k--', lw=1)
axes[0].set_title('ROC Curve'); axes[0].set_xlabel('FPR'); axes[0].set_ylabel('TPR')
axes[0].legend(fontsize=9)

baseline = y_test.mean()
axes[1].axhline(baseline, ls='--', color='grey', lw=1, label=f'Baseline={baseline:.3f}')
axes[1].set_title('Precision-Recall Curve')
axes[1].set_xlabel('Recall'); axes[1].set_ylabel('Precision')
axes[1].legend(fontsize=9)

plt.tight_layout(); plt.show()
```

```python
# Permutation importance (lebih andal — model-agnostik)
print('Menghitung permutation importance (Random Forest)...')
perm = permutation_importance(
    rf_model, X_test, y_test,
    scoring='f1', n_repeats=15, random_state=RS, n_jobs=-1
)
perm_df = pd.Series(perm.importances_mean, index=feat_cols).sort_values()

fig, ax = plt.subplots(figsize=(9, 7))
err = perm.importances_std[np.argsort(perm.importances_mean)]
colors_p = ['#FF6B6B' if v > 0 else '#A8A8A8' for v in perm_df]
ax.barh(perm_df.index, perm_df.values, xerr=err,
        color=colors_p, edgecolor='white', capsize=3)
ax.axvline(0, color='black', lw=0.8)
ax.set_title('Permutation Importance — Random Forest\n(Merah = berpengaruh positif terhadap F1)')
ax.set_xlabel('Mean Penurunan F1 saat Fitur Diacak')
plt.tight_layout(); plt.show()

print('Top 5 fitur paling penting (RF Permutation):')
print(perm_df.sort_values(ascending=False).head(5).round(4).to_string())
```
